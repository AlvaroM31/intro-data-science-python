## Unificación de Tablas

```python
for tabla in lista_tablas:
  print(f'N® de filas {len(tabla)}')
```

```python
ventas_completas = pd.concat(lista_tablas)
```

```python
ventas_completas.info()
#Total de filas son 1195905
```

```python
tiendas.info()
```

```python
productos.info()
```

```python
# Unir con los datos de productos y tiendas
ventas_full = pd.merge(ventas_completas,productos,on='id_prod').merge(tiendas, on ='id_tda')
```

## creando columnas adicionales

```python
ventas_full['monto'] = ventas_full['precio'] * ventas_full['cantidad']
```

```python
ventas_full['fecha'] = pd.to_datetime(ventas_full['fecha'])
ventas_full.info()
# Como string/objeto, es texto: solo puedes ordenarlo o compararlo como letras.
# Como datetime, es una fecha real: puedes hacer operaciones de calendario y
# ordenarlo de forma cronológica sin importar el formato del texto.
```

## creación de reportes pivot_table

```python
tabla1 = ventas_full.pivot_table(index='nombre_tda',columns='nombre_p',values='monto',aggfunc='sum',margins = True)
tabla1
```

```python
tabla2 = ventas_full.pivot_table(index='nombre_tda',columns='nombre_p',values='monto',aggfunc='sum')
tabla2
```

### Reportes por años y meses

```python
ventas_full['año'] = ventas_full['fecha'].dt.year
```

```python
ventas_full['mes'] = ventas_full['fecha'].dt.month_name() #puede ser solo month y saldria numero del 1 al 12
```

```python
tabla3 = ventas_full.pivot_table(index='nombre_tda',columns='año',values='monto',aggfunc='sum')
tabla3
```

```python
tabla4 = ventas_full.pivot_table(index='nombre_tda',columns='mes',values='monto',aggfunc='sum')
tabla4
```

Creación de la columna mes_dia

```python
ventas_full['mes_dia'] = ventas_full.fecha.dt.strftime("%m-%d")
```

Creación de la columna año_mes

```python
ventas_full['mes'] = ventas_full.fecha.dt.strftime("%Y-%m")
```

Creación de la tabla 5, que tendra un filtro de periodos menores al mes_dia < 06=25 (solo contara en cada año hasta el 25 de junio, no vale de julio a diciembre en esta tabla)

```python
tabla_5 = ventas_full.query("mes_dia <= '06-25'").pivot_table(index='nombre_tda',columns='año',values='monto',aggfunc='sum')
tabla_5
```

```python
tabla_5.T.plot(kind = 'bar', figsize=(12,4))
```

```python
series = ventas_full.pivot_table(index='mes',columns='ciudad',values='monto',aggfunc='sum')
series.plot(ylim=(0)) # x es el indice, y es el values y columns sera leyenda

# ylim significa que empezara la columna y, desde 0 en este casp
```

```python
series[1:-1].plot(ylim=(0)) #quitamos los valores que no estan completos los dias, en este caso seria el primero y ultimo
```

Otra forma de realizar el grafico eliminando esas series que no estan completas

```python
series[1:][:-1].plot(ylim=(0))
```

Si quiero un analisis de la grafica, mucho mas reciente, por ejemplo ventas de los ultimos 200 días, puedo realizar lo siguiente

```python
ventas_diarias = ventas_full.pivot_table(index ='fecha',values='monto',columns='ciudad',aggfunc='sum')[-200:]
```

```python
ventas_diarias.plot(figsize=(12,4),ylim=(60000,170000))
```

Creación de un promedio movil, que sea 14
