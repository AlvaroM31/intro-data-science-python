## Creacion de otro reporte

### Uniendo tablas

```python
ventas_full = ventas.merge(productos, on='id_prod').merge(tienda, on='id_tda')
```

Creando tabla

```python
ventas_full['monto'] = ventas_full['cantidad'] * ventas_full['precio']
```

```python
ventas_full
```

Generando reportes

```python
# Reportes
pivot_tabla = pd.pivot_table(ventas_full, values='monto', index='nombre_p',
                             columns='nombre_tienda', aggfunc='sum')
ventas_producto = pd.pivot_table(ventas_full, values='monto', index='nombre_p',
                                 aggfunc='sum')
ventas_tienda = pd.pivot_table(ventas_full, values='monto', index='nombre_tienda',
                               aggfunc='sum')
```

Guardando los reportes

```python
with pd.ExcelWriter(f'reporte2.xlsx') as writer:
    pivot_tabla.to_excel(writer, sheet_name='Ventas Producto-Tienda')
    ventas_producto.to_excel(writer, sheet_name='Ventas por Producto')
    ventas_tienda.to_excel(writer, sheet_name='Ventas por Tienda')
```

# Simplificando el reporte en 7 pasos

```python
import pandas as pd
```

```python
ventas = pd.read_excel("ventas.xlsx")
```

```python
ingresos = ventas.cantidad * ventas.precio
```

```python
ventas['ingresos'] = ingresos
```

```python
ventas['mes'] = ventas.fecha.dt.strftime("%Y-%m")
```

```python
tienda = pd.read_excel("tienda.xlsx")
ventas2 = pd.merge(tienda, ventas, on = 'id_tda')
productos = pd.read_excel("producto.xlsx")
ventas3 = pd.merge(ventas2, productos,  on = "id_prod")
```

```python
reporte1 = ventas3.pivot_table(index='ciudad',values='ingresos',columns='nombre_p',aggfunc='sum',margins=True)
reporte_arq = ventas3.query(" ciudad == 'Arequipa' ").pivot_table(index='nombre_p',values='ingresos',columns='mes',aggfunc='sum')
reporte_lim = ventas3.query(" ciudad == 'Lima' ").pivot_table(index='nombre_p',values='ingresos',columns='mes',aggfunc='sum')
reporte_cus = ventas3.query(" ciudad == 'Cusco' ").pivot_table(index='nombre_p',values='ingresos',columns='mes',aggfunc='sum')
```

```python
with pd.ExcelWriter(f"reporte.xlsx") as writer:
  reporte1.to_excel(writer, sheet_name='Reporte_Prod_Tienda')
  reporte_arq.to_excel(writer, sheet_name='Reporte_arequipa')
  reporte_cus.to_excel(writer, sheet_name='Reporte_cusco')
  reporte_lim.to_excel(writer, sheet_name='Reporte_lima')
```
