FORMA PRÁCTICA DE UNIR ESTOS CONCEPTOS: Quiero saber la tasa de crecimiento de las ventas:

```python
ventas = [108,124,139,157,184,210,265,300]
g = (ventas[-1]/ventas[0])**(1/(len(ventas)-1))-1
g
```

Esta bien de esta manera, pero también es bueno dividir los pasos:

```python
Y_t = ventas[-1]
Y_1 = ventas[0]
N_crec = len(ventas)-1
G= (Y_t/Y_1)**(1/N_crec)-1
G
```

Ahora tenemos la tasa de crecimiento G así que a partir del periodo t, podemos pronosticar, por ejemplo:

```python
steps = int(input('Ingrese el numero de periodos\n'))
forecast = Y_t*(1+g)**steps
print(f'La proyeccion para el periodo {steps} es {forecast:.1f}')
```

### Bucle range

son iteraciones que permiten automatizar ciertos programas

**Ingresos[1]** o **ingresos[2]**, me permite cambiar de valores entre la lista ingresos, en este caso, porque no consideramos ¿ingresos[i]?, y que este se itere todas las veces.

```python
for i in range(4):
  print(i)
```

```python
cantidad = len(ingresos)
for i in range(cantidad):
  print(ingresos[i])
```

En los bucle for, también puedo saltar 1 numero, es decir que vaya la lista en 2 en 2; de la siguiente manera:

```python
for i in range(0,cantidad,2):
  print(ingresos[i])
```

Tambien puede ir de forma negativa, por ejemplo disminuyendo de 2 en 2

```python
for i in range(cantidad-1,-1,-2):
  print(ingresos[i])
```

Casuística: Quiero saber cuanto de valorización total tendré cada mes en el futuro

```python
capital = 1000
interes = 0.05
meses = int(input('¿Dime cuantos meses dejarás tu capital en el banco?\n'))
for i in range(1,meses+1):
  capital = capital * (1+interes)
  print(f'En un periodo de {i} meses se obtendra una valorización de {capital:.1f}')
```

Genial, pero ahora me gustaria guardarlo en una lista esos valores de capitalización

```python
capital = 1000
interes = 0.05
lista_valorización = []
meses = int(input('¿Dime cuantos meses dejarás tu capital en el banco?\n'))
for i in range(1,meses+1):
  capital = capital * (1+interes)
  capital = round(capital,1)
  lista_valorización.append(capital)
  print(f'En un periodo de {i} meses se obtendra una valorización de {capital:.1f}')
```

Ejercicio

Programa que le pregunte al usuario cuantos periodos quiere registrar y luego le pregunte periodo a periodo el valor de las ventas.

```python
Lista = []
n = int(input('¿Cuántos periodos desea ingresar?\n'))
for x in range(1,n+1):
  ventas = int(input(f'¿Cuál es el valor de las ventas en el periodo {x}?\n'))
  print(f'En el periodo: {x}, el valor de ventas es: {ventas}')
  Lista.append(ventas)
print(f'Las ventas registradas son: \n{Lista}')
input('Seleccione la celda enter para hallar la tasa de crecimiento\n')
vi = Lista[0]
vf = Lista[-1]
n = len(Lista)-1
g = (vf/vi)**(1/n) - 1
Lista = list(zip(range(1,n+2),Lista))
print(f"La tasa de crecimiento {(g*100):.1f}%")
```

### Método While:

Ejecuta una cantidad indefinida de repeticiones en un mismo bucle, hasta que se cumpla un criterio.

```python
x = 2
while x < 4:
  print(x)
  x = x+1
```

EJEMPLO: Si las ventas actuales son 1500 y la meta es 2500, ¿cuantos meses faltan? si la tasa de crecimiento mensual es 8%.

```python
ventas_actuales = 1500
meta = 2500
n = 0
crecimiento = 0.08
while meta > ventas_actuales:
  n = n + 1
  ventas_actuales = round((ventas_actuales * (1+crecimiento)**n),1)
  print(f'Van {ventas_actuales} ventas actuales en el periodo {n}')
print(f'Se llego a la meta en el periodo {n} con {ventas_actuales} ventas')
```

EJEMPLO: Genere un programa en donde le pregunte al usuario si esta lloviendo, si el usuario responde algo distinto a si o no(en sus diferentes versiones si, SI, No,...) repreguntar. Si responde bien dar las gracias.

```python
resp = input('¿Esta lloviendo?\nSolo puede ingresar [si o no]\n').strip().lower()
while resp not in ['si', 'no']:
  resp = input('¿Esta lloviendo?\nSolo puede ingresar [si o no]\n').strip().lower()
print(f'La respuesta es:\n{resp}')
```

Otra forma de realizar este mismo ejercicio

```python
while True:
   respuesta = input("¿Está lloviendo? (Sí/No): \n").strip().lower()
   if respuesta == "si":
      print("¡Gracias por tu respuesta!")
      break
   elif respuesta == "no":
      print("¡Gracias por tu respuesta!")
      break
   else:
      print("Por favor, responde con 'Sí' o 'No'. Intenta de nuevo.")
```
